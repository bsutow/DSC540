```python
#Shows no n/a values#
for c in adult.columns:
    miss=adult[c].isnull().sum()
    if miss>0:
        print("This has {} missing values".format(c,miss))
    else:
        print("This has NO missing values@".format(c))
```

```python
#Creates new data frame with three variable, and shows the values of each descriptive statistic below#
newadult= adult.loc[[i for i in range(0,10)], ['age','education','occupation'] ]
newadult
```

```python
#Creates filtered data set showing between ages 30 and 50#
adultfiltered=newadult[(newadult['age']>=30) & (newadult['age']<=50)]
adultfiltered
```

```python
#Groups education by age#
gbed=newadult.groupby(['education']).mean()['age']
gbed
```

```python
#Groups occupation by age#
gbocc = newadult.groupby(['occupation']).describe()['age']
gbocc
```

```python
#Creates bar chart with gbocc sshowing the counts of positions#
plt.bar(x=gbocc.index, height=gbocc['count']), plt.xticks(rotation=90)
```

```python
#Merges filtered data with gbocc grouping#
pd.merge(adultfiltered, gbocc, on = ['occupation'], how= 'inner')
```

```python
#Merges filtered data with gbed grouping#
pd.merge(adultfiltered, gbed, on = ['education'], how= 'inner')
```
